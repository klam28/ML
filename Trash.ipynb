```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
```

```python
# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :2] # we only take the first two features. We could
y = iris.target
```

```python
# we create an instance of SVM and fit out data. We do not scale our
# data since we want to plot the support vectors
C = 1.0 # SVM regularization parameter/Cost Parameter
# A.k.a penalty parameter or the error term, which controls
# for between smooth decision boundary and classifying training
# points correctly


# Linear does well where N > 1000
# svc = svm.SVC(kernel='linear', C=1,gamma='auto').fit(X, y)

# Gamma = Learning Rate ~ Higher Gamma will try to exact fit training data, can lead to over-fitting
svc = svm.SVC(kernel='rbf', C=1,gamma='auto').fit(X, y)
```

```python
# create a mesh to plot in
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

h = (x_max / x_min)/100
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
 np.arange(y_min, y_max, h))
```

```python
plt.subplot(1, 1, 1)

Z = svc.predict(np.c_[xx.ravel(), yy.ravel()]) # Predict the Space For the Points 
Z = Z.reshape(xx.shape)


plt.contourf(xx, yy, Z, cmap=plt.cm.winter, alpha=0.3)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.winter, alpha = 1)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('SVC with linear kernel')
plt.show()
```
